```python
import pandas as pd
filepath = "/content/slang_dataset.csv"
df = pd.read_csv(filepath)

# Define all keyword lists (same as previous)
finance_keywords = ["mari", "mula", "ndai", "bag", "maObama", "float", "shagi"]
sports_keywords = ["bhora", "muzhalo", "netball", "muZhalo", "cr7", "zimwarriors"]
farewell_keywords = ['zvakare', 'bye', 'laters', 'ndakufamba', 'ndaenda', 'stay blessed', 'keep well', 'tobatana', 'toonana', 'ndokubata']
religion_keywords = ['amen', 'grace', 'Jehova', 'blessed', 'ndinonyengetera', 'prophet', 'anointed', 'madzibaba', 'postori',
                     'mufundisi', 'muvhangeri', 'muparidzi', 'denga', 'hell', 'gehena', 'gava', 'tendeuka', 'hallelujah',
                     'verse', 'Jesus', 'Satan', 'demon', 'mweya', 'paseka', 'matumba', 'pentecost']
support_keywords = ['ndibatsirei', 'ma1', 'ndarasa', 'ari kundishungurudza', 'ndodii', 'ndikuda', 'rubatsiro', 'ndipe']
news_keywords = ['breaking', 'guhwa', 'varikuroora', 'hanzi', 'heard', 'trending', 'nhema', 'rumor', 'latest', 'nyaya']
location_keywords = ['paJulius', 'paChikwanha', 'kumagumaguma', 'pa corner', 'mupurazi', 'CBD', 'base', 'hood', 'tonaz', 'roots', 'den', 'fazh']
health_keywords = ['bp', 'sick', 'clinic', 'ngomwa', 'ma ARVs', 'rwara', 'mental', 'gula']
school_keywords = ['maexam', 'deadline', 'results', 'fazh', 'makwikwi', 'GZU', 'UZ', 'msu']
romance_keywords = ['bae', 'sweety', 'ndokuda', 'ndine crush', 'moyo']
complaint_keywords = ['ma1', 'zvakaoma', 'ngavaende', 'crisis']
gratitude_keywords = ['bless up', 'respect', 'madam boss', 'mwaaa']

def get_intent(message):
    # Convert message to string to handle non-string types
    message = str(message)
    message = message.lower()
    words = message.split()  # Tokenize message

    for word in words:
        if word in finance_keywords:
            return "finance"
        elif word in sports_keywords:
            return "sports"
        elif word in farewell_keywords:
            return "farewell"
        elif word in religion_keywords:
            return "religion"
        elif word in support_keywords:
            return "help_request"
        elif word in news_keywords:
            return "news"
        elif word in location_keywords:
            return "location"
        elif word in health_keywords:
            return "health"
        elif word in school_keywords:
            return "education"
        elif word in romance_keywords:
            return "romance"
        elif word in complaint_keywords:
            return "complaint"
        elif word in gratitude_keywords:
            return "gratitude"

    return "chitchat"  # fallback

# Apply function to your dataset
df["intent"] = df["message"].apply(get_intent)

df.head()
```

```python
import pandas as pd
from datasets import Dataset # Import the Dataset class
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments

filepath = '/content/slang_dataset_with_intent.csv'
df = pd.read_csv(filepath)

# Reduce the training data size by sampling
df = df.sample(frac=0.1, random_state=42) # Sample 10% of the data, use a random_state for reproducibility

# Convert pandas DataFrame to Hugging Face Dataset
dataset = Dataset.from_pandas(df)

tokenizer = AutoTokenizer.from_pretrained("xlm-roberta-base")

# Map labels to integers and define number_of_labels
unique_intents = df['intent'].unique().tolist()
label_map = {intent: i for i, intent in enumerate(unique_intents)}
number_of_labels = len(unique_intents)

# Add a 'labels' column with integer-encoded labels
dataset = dataset.map(lambda examples: {'labels': [label_map[intent] for intent in examples['intent']]}, batched=True)


def tokenize_function(examples):
    # Ensure 'message' column exists and is used for tokenization
    # Explicitly convert each item in the batch to string
    messages = [str(msg) for msg in examples["message"]]
    return tokenizer(messages, truncation=True, padding="max_length")

tokenized_dataset = dataset.map(tokenize_function, batched=True)

model = AutoModelForSequenceClassification.from_pretrained("xlm-roberta-base", num_labels=number_of_labels)

training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch", # Corrected argument name
    learning_rate=2e-5,
    per_device_train_batch_size=8, # Reduced batch size
    num_train_epochs=3,
    weight_decay=0.01,
    report_to="none"  # Disable wandb integration
)

# You'll likely need to split your dataset into train and evaluation sets
# For now, I'll use the full dataset for both, but you should split it properly
# Example:
# train_testsplit = tokenized_dataset.train_test_split(test_size=0.2)
# train_dataset = train_testsplit['train']
# eval_dataset = train_testsplit['test']


trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset, # Use the full dataset for training (consider splitting)
    eval_dataset=tokenized_dataset,   # Use the full dataset for evaluation (consider splitting)
)

trainer.train()
```

```python
# prompt: save this model

# Define the directory to save the model
output_dir = "./finetuned_intent_model"

# Save the model
trainer.save_model(output_dir)

print(f"Model saved to {output_dir}")
```
