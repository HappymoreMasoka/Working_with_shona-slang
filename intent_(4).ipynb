```python
import pandas as pd
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from transformers import EarlyStoppingCallback  # Import EarlyStoppingCallback directly from transformers

# Load and sample data
filepath = '/content/slang_dataset_with_intent.csv'
df = pd.read_csv(filepath)
df = df.sample(frac=0.1, random_state=42)  # Sample 10% of the data

# Convert pandas DataFrame to Hugging Face Dataset
dataset = Dataset.from_pandas(df)

# Initialize tokenizer
tokenizer = AutoTokenizer.from_pretrained("xlm-roberta-base")

# Map labels to integers
unique_intents = df['intent'].unique().tolist()
label_map = {intent: i for i, intent in enumerate(unique_intents)}
number_of_labels = len(unique_intents)

# Add a 'labels' column with integer-encoded labels
dataset = dataset.map(lambda examples: {'labels': [label_map[intent] for intent in examples['intent']]}, batched=True)

# Tokenize function
def tokenize_function(examples):
    messages = [str(msg) for msg in examples["message"]]
    return tokenizer(messages, truncation=True, padding="max_length")

tokenized_dataset = dataset.map(tokenize_function, batched=True)

# Split dataset into train and validation sets
train_testsplit = tokenized_dataset.train_test_split(test_size=0.2, seed=42)  # Proper train-val split
train_dataset = train_testsplit['train']
eval_dataset = train_testsplit['test']

# Initialize model with increased dropout
model = AutoModelForSequenceClassification.from_pretrained(
    "xlm-roberta-base",
    num_labels=number_of_labels,
    hidden_dropout_prob=0.3,  # Increased dropout (default is 0.1)
    attention_probs_dropout_prob=0.3  # Increased attention dropout
)

# Define training arguments with increased weight decay
training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    save_strategy="epoch", # Set save strategy to epoch to match eval strategy
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,  # Kept at 3, early stopping will handle if needed
    weight_decay=0.1,  # Increased from 0.01 to 0.1 for stronger regularization
    report_to="none",
    load_best_model_at_end=True,  # Load the best model based on validation loss
    metric_for_best_model="eval_loss"  # Use validation loss to determine best model
)

# Initialize trainer with early stopping
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=2)]  # Stop if no improvement after 2 evals
)

# Train the model
trainer.train()
```

```python
intent_to_response = {
    "greeting": "Mhoro! Unoda rubatsiro nei nhasi?",
    "ask_fees": "Mari yacho inotsamira neprogram yaunosarudza. Unoda kuziva nezvei chaizvo?",
    "admissions": "Kuti uwane admission, unofanirwa kuzadza application form uye wotumira marequirements.",
    "location": "University iri muNew York, USA.",
    "deadline": "Deadline yeapplication ndeye November 30, 2025.",
    # Add your actual intents and responses here
}

def classify_intent(text):
    result = classifier(text)
    pred = max(result[0], key=lambda x: x["score"])
    return pred["label"], pred["score"]

def chatbot_response(user_input):
    intent, confidence = classify_intent(user_input)
    response = intent_to_response.get(intent, "Ndine urombo, handina kunzwisisa. Unogona kudzokorora here?")
    return f"[{intent} | confidence: {confidence:.2f}] {response}"

print("Shona Chatbot 🤖: Mhoro! Ndibvunzei zvamunoda (type 'exit' to quit)")

while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Zvakanaka, tichaonana zvakare!")
        break

    response = chatbot_response(user_input)
    print("Chatbot:", response)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

model_name = "HuggingFaceH4/zephyr-7b-beta"  # You can switch to other Instruct models
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto")

generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

# Few-shot prompt
def classify_intent_few_shot(message):
    prompt = f"""
You are a helpful assistant that classifies Shona messages into predefined intents.

Examples:
Message: "mhoro" → Intent: greeting
Message: "ndinoda kuziva mafees" → Intent: ask_fees
Message: "ndinoda kunyorera admission" → Intent: admissions
Message: "unogara kupi university?" → Intent: location
Message: "deadline yekunyoresa iri riinhi?" → Intent: deadline

Now classify:
Message: "{message}" → Intent:"""

    output = generator(prompt, max_new_tokens=10, do_sample=False)[0]['generated_text']

    # Extract intent from generated output
    intent = output.split("→ Intent:")[-1].strip().split("\n")[0]
    return intent
```
