```python
def detect_contexts(message):
    msg = message.lower()
    contexts = []

    if any(word in msg for word in ["mom", "amai", "mwana", "sisi", "bro",'mukoma','mwana','hama','mkanya', 'mrewa','dhara','sekuru','asekuru', 'mudhara','sistren', 'hanzvadzi', 'dad', 'father', 'gogo', 'gulez', 'ninez', 'chembere','mukadzi','muninina', 'mufana', 'mufana', 'murora', 'tezvara', 'ambuya','gotwe', 'dangwe', 'tete', 'tsano', 'xaaa','sha', 'shamwari','murongo','blaz','mhamha', 'baba', 'amai', 'ba', 'mukwasha']):
        contexts.append("family")
    if any(word in msg for word in ["mwari", "grace", "bless", "amen", "saints", "god",'amen', 'grace', 'Jehova', 'blessed', 'ndinonyengetera', 'prophet', 'anointed', 'madzibaba', 'postori', 'mufundisi', 'muvhangeri','muparidzi','denga', 'hell','gehena', 'gava','tendeuka','hallelujah','verse','Jesus','Satan','demon','mweya', 'paseka','matumba','pentecost' ]):
        contexts.append("religion")
    if any(word in msg for word in ["wangu", "wanguda", "chamhamha", "changu", 'bro', 'sis', 'xaa', 'shamwari',' sha','sahwira', 'fam','boys', 'guys', 'team','majita', 'mufesi','sazi', 'sahwi','']):
        contexts.append("friendship")

    if any(word in msg for word in ["thank", "thanx", "thanks", "ndatenda", 'maita basa', 'ndinotenda','tenda', 'ndatenda', 'tatenda','wasimbisa','wagona','wakaipa','warema','wapedza', 'manake']):
        contexts.append("appreciation")
    if any(word in msg for word in ["happy mothers day", "congrats", "makorokoto", 'birthday','bhavhadheyi',' fathers','congratulations', 'mothers']):
        contexts.append("celebration")
    if any(word in msg for word in ["help", "ndibatsireiwo", "pliz",'jega','ndipo', 'ndipeiwo','ndokumbirawo', 'kumbirawo', 'kumbira','dambudziko']):
        contexts.append("requests")

    if any(word in msg for word in ["sarai", "bye", "ndokuonai", "toonana", "mushe", "tobatana"]):
        contexts.append("farewell")

    if any(word in msg for word in ["fazh", "school", "teacher", "head", 'dzidza','chikoro', 'zvikoro','maexam', 'deadline', 'results', 'fazh', 'makwikwi', 'GZU', 'UZ', 'msu']):
        contexts.append("educational")
    if any(word in msg for word in ["bae", "babe", "love", 'ndokuda', 'sweetie','hun']):
        contexts.append("romantic")
    if any(word in msg for word in ["shanda", "basa", "work", 'graft', 'mabasa']):
        contexts.append("work")
        if any(word in msg for word in ["hahaha", "kkk", "joke"]):
        contexts.append("humor")
    if any(word in msg for word in ["putseki", "kushata", "miswa"]):
        contexts.append("argument")
    if any(word in msg for word in ['bhora', 'netball', 'muZhalo', 'cr7', 'zimwarriors']):
        contexts.append("sports")
    if any(word in msg for word in ['paJulius', 'paChikwanha', 'kumagumaguma', 'pa corner', 'mupurazi', 'CBD', 'base', 'hood','tonaz', 'roots','den','fazh']):
        contexts.append("location")

    if any(word in msg for word in ["sorry", "nematambudziko", "nyaradzwayi", "hurombo"]):
        contexts.append("sympathy")
    if any(word in msg for word in ["sick", "rwara", "flue", 'hospital','clinic']):
        contexts.append("health")
    if any(word in msg for word in ["help", "ndibatsireiwo", "pliz"]):
        contexts.append("politics")

    if not contexts:
        contexts.append("GEN")

    return contexts

# Example
msg = "Good morning saints, God is surely good"
print(detect_contexts(msg))
# Output: ['REL', 'GEN']
```
